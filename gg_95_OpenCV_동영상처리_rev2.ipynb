```python
import cv2
from google.colab import drive
# from pytube import YouTube
from pytubefix import YouTube
```

```python
# Youtube 객체 생성 (제목, 길이, 조회수, 스트림 정보 등 메타 데이터 포함)
yt = YouTube("https://www.youtube.com/watch?v=OwJPPaEyqhI")

# 스트림 필터링 (영상과 오디오가 합쳐진 스트림만 선택, 가장 높은 해상도)
yt = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
yt.download('/content/')
```

```python
# 비디오 경로 지정
# file name 을 못 읽으면 New_Face.mp4 로 수정할 것
video_path = "/content/New_Face.mp4"
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
video = cv2.VideoCapture(video_path) # 비디오 파일 열기
fps = video.get(cv2.CAP_PROP_FPS)    # 비디오의 초당 프레임 수(FPS, Frames Per Second)를 가져오는 함수
fps
```

```python
frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
print("총 프레임 수:", frame_count)
duration = frame_count / fps                     # 영상 길이(초)
print("FPS:", fps)
print("총 프레임 수:", frame_count)
print("영상 길이(초):", duration)
```

```python
import os
video = cv2.VideoCapture(video_path)
success = True
count = 1

while success:
  success,frame = video.read()
  if not os.path.exists("/content/test"):
      os.mkdir('test')
  name = '/content/test/'+str(count)+'.jpg'

  if success != True:
    print("failed to read YouTube video.")
    break

  else:
    count += 1
    if count % 100 == 0:  # 100 frame 마다 하나씩
        cv2.imwrite(name, frame)
        print('Frame {} extracted.'.format(count))

        if (count == 2000):  # 20 frame 캡처
          break

print("20 frames extracted successfully")
```

- DASH 관련 Stream을 제거하려면 Progressive = True (progressive mode 에서의 최대 해상도는 720 p)
